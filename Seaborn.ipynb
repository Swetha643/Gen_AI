```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
mtcars=pd.read_csv(io.BytesIO(uploaded['mtcars.csv']))
```

```python
mtcars.head()
```

```python
mtcars.info()
```

```python
mtcars.shape #32 rows and 12 column
```

# Bar Plot

```python
res=sns.barplot(mtcars['cyl'],mtcars['mpg'])
```

```python
res=sns.barplot(x='cyl' ,y='mpg',data=mtcars)
plt.show()
```

# color

```python
res=sns.barplot(x='cyl' ,y='mpg',data=mtcars,color="hotpink")
plt.show()
```

# Palette

```python
res=sns.barplot(x='am' ,y='mpg',data=mtcars,palette="Set2")
plt.show()
```

```python
res=sns.barplot(x='gear' ,y='hp',data=mtcars,palette="rocket")
plt.show()
```

# count plot

```python
res=sns.countplot(x='cyl',data=mtcars,palette="Set1")
```

```python
res=sns.countplot(x='carb',data=mtcars,palette="magma")
```

# Horizontal bar

```python
res=sns.countplot(y='gear',data=mtcars,palette="magma")
```

# Hue parameter - group plot for colour encoding

```python
res=sns.countplot(x='gear',hue='cyl',data=mtcars,palette="magma")
```

# Displot(distribution plot)

```python
sns.distplot(mtcars.mpg,bins=10,color='g')
```

# Heatmap

```python
sns.heatmap(mtcars.corr(), cbar=True,linewidth=0.5)
```

# scatterplot

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].value_counts()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length' ,data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species')
```

# size parameter

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species',size='species')
```

# pairplot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette="Set1")
```

# lmplot

```python
sns.lmplot(x='petal_length',y='petal_width',data=iris)
```

```python
sns.lmplot(x='petal_length',y='petal_width',data=iris,hue='species',markers=['o','*','^'])
```

# boxplot() function

```python
sns.boxplot(x='species',y='sepal_width',data=iris)
```
